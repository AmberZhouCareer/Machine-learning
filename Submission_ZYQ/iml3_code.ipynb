```python
#load data
import pandas as pd

dev_count = pd.read_csv(open("dev_count.csv"))
dev_full = pd.read_csv(open("dev_full.csv"))
dev_glove = pd.read_csv(open("dev_glove.csv"))
dev_tfidf = pd.read_csv(open("dev_tfidf.csv"))

test_count = pd.read_csv(open("test_count.csv"))
test_full = pd.read_csv(open("test_full.csv"))
test_glove = pd.read_csv(open("test_glove.csv"))
test_tfidf = pd.read_csv(open("test_tfidf.csv"))

train_count = pd.read_csv(open("train_count.csv"))
train_full = pd.read_csv(open("train_full.csv"))
train_glove = pd.read_csv(open("train_glove.csv"))
train_tfidf = pd.read_csv(open("train_tfidf.csv"))
```

```python
#Extract labels
train_label = train_count["sentiment"]
dev_label = dev_count["sentiment"]

#Extract features
dev_count_features = dev_count["tweet"]
dev_full_features = dev_full["tweet"]
dev_glove_features = dev_glove["tweet"]
dev_tfidf_features = dev_tfidf["tweet"]

test_count_features = test_count["tweet"]
test_full_features = test_full["tweet"]
test_glove_features = test_glove["tweet"]
test_tfidf_features = test_tfidf["tweet"]

train_count_features = train_count["tweet"]
train_full_features = train_full["tweet"]
train_glove_features = train_glove["tweet"]
train_tfidf_features = train_tfidf["tweet"]
```

```python
#data processing
vocab = open("vocab.txt", "r")
line_num = 0
while vocab.readline():
    line_num += 1

def data_processing(features):
    result = []
    for i in range(len(features)):
        temp = [0] * line_num
        for word, times in eval(features[i]):
            temp[word] = times
        result.append(temp)
    return result

train_count_input = data_processing(train_count_features)
train_tfidf_input = data_processing(train_tfidf_features)
test_count_input = data_processing(test_count_features)
test_tfidf_input = data_processing(test_tfidf_features)
dev_count_input = data_processing(dev_count_features)
dev_tfidf_input = data_processing(dev_tfidf_features)
```

```python
def data_processing_glove(glove):
    result = []
    for i in glove:
        result.append(eval(i))
    return result
train_glove_input = data_processing_glove(train_glove_features)
test_glove_input = data_processing_glove(test_glove_features)
dev_glove_input = data_processing_glove(dev_glove_features)
```

```python
# Naive Bayes
from sklearn.naive_bayes import BernoulliNB
def BNB(trainX, trainY):
    BNB = BernoulliNB()
    BNB.fit(trainX, trainY)
    return BNB

from sklearn.naive_bayes import MultinomialNB
def MNB(trainX, trainY):
    MNB = MultinomialNB()
    MNB.fit(trainX, trainY)
    return MNB

from sklearn.naive_bayes import GaussianNB
def GNB(trainX, trainY):
    GNB = GaussianNB()
    GNB.fit(trainX, trainY)
    return GNB

# Logistic Regression
from sklearn.linear_model import LogisticRegression
def LR(trainX, trainY):
    LR = LogisticRegression(solver='lbfgs', max_iter=100000)
    LR.fit(trainX, trainY)
    return LR

# Neural Network
from sklearn.neural_network import MLPClassifier
def NN(trainX, trainY):
    NN = MLPClassifier(max_iter=1000)
    NN.fit(trainX, trainY)
    return NN

# Decision Trees
from sklearn.tree import DecisionTreeClassifier
def DT(trainX, trainY):
    DT = DecisionTreeClassifier(criterion='gini')  
    DT.fit(trainX, trainY)
    return DT
```

```python
# Naive Bayes
BNB_count = BNB(train_count_input, train_label)
BNB_tfidf = BNB(train_tfidf_input, train_label)
BNB_glove = BNB(train_glove_input, train_label)

MNB_count = MNB(train_count_input, train_label)
MNB_tfidf = MNB(train_tfidf_input, train_label)

GNB_count = GNB(train_count_input, train_label)
GNB_tfidf = GNB(train_tfidf_input, train_label)
GNB_glove = GNB(train_glove_input, train_label)
```

```python
# Logistic Regression
LR_count = LR(train_count_input, train_label)
LR_tfidf = LR(train_tfidf_input, train_label)
LR_glove = LR(train_glove_input, train_label)
```

```python
# Neural Network
NN_count = NN(train_count_input, train_label)
NN_tfidf = NN(train_tfidf_input, train_label)
NN_glove = NN(train_glove_input, train_label)
```

```python
# Decision Trees
DT_count = DT(train_count_input, train_label)
DT_tfidf = DT(train_tfidf_input, train_label)
DT_glove = DT(train_glove_input, train_label)
```

```python
def predict(model, test):
    return model.predict(test)

# Naive Bayes
BNB_count_predict = predict(BNB_count, dev_count_input)
BNB_tfidf_predict = predict(BNB_tfidf, dev_tfidf_input)
BNB_glove_predict = predict(BNB_glove, dev_glove_input)

MNB_count_predict = predict(MNB_count, dev_count_input)
MNB_tfidf_predict = predict(MNB_tfidf, dev_tfidf_input)

GNB_count_predict = predict(GNB_count, dev_count_input)
GNB_tfidf_predict = predict(GNB_tfidf, dev_tfidf_input)
GNB_glove_predict = predict(GNB_glove, dev_glove_input)

# Logistic Regression
LR_count_predict = predict(LR_count, dev_count_input)
LR_tfidf_predict = predict(LR_tfidf, dev_tfidf_input)
LR_glove_predict = predict(LR_glove, dev_glove_input)

# Neural Network
NN_count_predict = predict(NN_count, dev_count_input)
NN_tfidf_predict = predict(NN_tfidf, dev_tfidf_input)
NN_glove_predict = predict(NN_glove, dev_glove_input)

# Decision Trees
DT_count_predict = predict(DT_count, dev_count_input)
DT_tfidf_predict = predict(DT_tfidf, dev_tfidf_input)
DT_glove_predict = predict(DT_glove, dev_glove_input)
```

```python
from sklearn.metrics import accuracy_score
def evaluate(Y_real, Y_predict):
    return accuracy_score(Y_real, Y_predict)

# Naive Bayes
print("BNB_count acc :", evaluate(dev_label, BNB_count_predict))
print("BNB_tfidf acc :", evaluate(dev_label, BNB_tfidf_predict))
print("BNB_glove acc :", evaluate(dev_label, BNB_glove_predict))

print("MNB_count acc :", evaluate(dev_label, MNB_count_predict))
print("MNB_tfidf acc :", evaluate(dev_label, MNB_tfidf_predict))

print("GNB_count acc :", evaluate(dev_label, GNB_count_predict))
print("GNB_tfidf acc :", evaluate(dev_label, GNB_tfidf_predict))
print("GNB_glove acc :", evaluate(dev_label, GNB_glove_predict))

# Logistic Regression
print("LR_count acc :", evaluate(dev_label, LR_count_predict))
print("LR_tfidf acc :", evaluate(dev_label, LR_tfidf_predict))
print("LR_glove acc :", evaluate(dev_label, LR_glove_predict))

# Neural Network
print("NN_count acc :", evaluate(dev_label, NN_count_predict))
print("NN_tfidf acc :", evaluate(dev_label, NN_tfidf_predict))
print("NN_glove acc :", evaluate(dev_label, NN_glove_predict))

# Decision Trees
print("DT_count acc :", evaluate(dev_label, DT_count_predict))
print("DT_tfidf acc :", evaluate(dev_label, DT_tfidf_predict))
print("DT_glove acc :", evaluate(dev_label, DT_glove_predict))
```
